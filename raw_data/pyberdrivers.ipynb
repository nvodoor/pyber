```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fileone = "city_data.csv"
filetwo = "ride_data.csv"
```

```python
f1 = pd.read_csv(fileone)
f2 = pd.read_csv(filetwo)
```

```python
f1
```

```python
f2
```

```python
def getfare(group):
    return {'ride_count': group.count(), 'avg': np.round(group.mean(),2)}
groupby = f2['fare'].groupby(f2['city']).apply(getfare).unstack()
groupby['city'] = groupby.index
# (group.sum()/group.count() --> if average means sum divided count
```

```python
groupby
```

```python
mergetableone = pd.merge(f1,f2, how="inner", on="city")
```

```python
mergetable = pd.merge(mergetableone, groupby, how="inner", on="city")
# mergetable = mergetable.drop_duplicates(subset=['ride_id'],keep=False)
```

```python
mergetable
```

```python
import matplotlib.patches as mpatches
fig = plt.figure(dpi=200)
ax = fig.add_axes([0.8,0.8,0.8,0.8])

drivers = mergetable['driver_count']
x_axis = np.arange(0, len(mergetable['city'].unique()))
avg = mergetable['avg'].value_counts()
y_axis = np.arange(0, len(mergetable['city'].unique()))

color_dict = {'Urban': 'blue', 'Suburban': 'green', 'Rural': 'yellow'}

ax.set_xlim(0, max(mergetable['ride_count']+2))
ax.set_ylim(0, max(avg))
ax.set_title('Pyber Ride Sharing Data (2017)')
ax.set_xlabel('Total Number of Rides Per City')
ax.set_ylabel('Avg Fare($)')
ax.scatter(mergetable['ride_count'], mergetable['avg'], color=[color_dict[i] for i in mergetable['type']], edgecolor='black', s=drivers)
blue_patch = mpatches.Patch(color='blue', label='Urban')
yellow_patch = mpatches.Patch(color='yellow', label='Rural')
green_patch = mpatches.Patch(color='green', label='Suburban')
plt.legend(handles=[blue_patch,green_patch,yellow_patch], title='City Type')
plt.style.use('ggplot')
plt.text(-16,0, " Note: The circle size \n represents how many drivers \n are in that city.")
plt.show()
```

```python
def getfareride(group):
    return {'ride_count': group.count(), 'fares': np.round(group.sum(),2)}
groupbyone = mergetableone['fare'].groupby(mergetableone['type']).apply(getfareride).unstack()
groupbyone['type'] = groupbyone.index
```

```python
groupbyone
```

```python
def getdrivers(group):
    return {'drivers': group.sum()}
groupbytwo = f1['driver_count'].groupby(f1['type']).apply(getdrivers).unstack()
groupbytwo['type'] = groupbytwo.index
```

```python
groupbytwo
```

```python
fig = plt.figure(dpi=80)
ax = fig.add_axes([1, 1, 1, 1])


colors = ['lightblue','lime','orange']
explode = (0,0,0.1)


ax.set_title("% of Total Fares By Type")
ax.pie(groupbyone['fares'], explode=explode, labels=groupbyone['type'], colors=colors, autopct = "%1.1f%%", shadow=True, startangle=120)
ax.axis("equal")

plt.show()
```

```python
fig = plt.figure(dpi=80)
axone = fig.add_axes([1, 1, 1, 1])


colors = ['lightblue','lime','orange']
explode = (0,0,0.1)


axone.set_title("% of Total Rides By Type")
axone.pie(groupbyone['ride_count'], explode=explode, labels=groupbyone['type'], colors=colors, autopct = "%1.1f%%", shadow=True, startangle=90)
axone.axis("equal")

plt.show()
```

```python
fig = plt.figure(dpi=80)
axtwo = fig.add_axes([1, 1, 1, 1])


colors = ['lightblue','lime','orange']
explode = (0,0,0.1)


axtwo.set_title("% of Total Drivers By Type")
axtwo.pie(groupbytwo['drivers'], explode=explode, labels=groupbytwo['type'], colors=colors, autopct = "%1.1f%%", shadow=True, startangle=150)
axtwo.axis("equal")

plt.show()
```
